```python
import multiprocessing
import numpy as np
import xarray as xr
import os
import pandas as pd
import netCDF4 as nc
import matplotlib as mpl
import matplotlib.pyplot as plt
import statsmodels.api as sm
from scipy.stats import pearsonr
from scipy.stats import linregress
from scipy.optimize import curve_fit
from pymer4.models import Lmer
import math
import seaborn as sns
```

```python
type_data = pd.read_csv('./Site_igbp.csv', encoding='latin-1')
site_IGBP = type_data['igbp'].values
site_lat = type_data['lat'].values
site_ID =  type_data['site'].values


path = 'D:/RECO_T_sensitivity/T_sensitivity_data/LMX_data_T_effect_result3_nighttime/'

site = ['site']
igbp = ['igbp']
T = ['T']
Q10 = ['Q10']
SE = ['SE']
p = ['pvalue']
sig = ['significance']
num = ['num']
for csv_file in os.listdir(path):
    
    for csv_file1 in os.listdir(path+csv_file+'/'):
        Ti = int(csv_file1[7:-4])
        seleceted_data = pd.read_csv(path+csv_file+'/'+csv_file1)
        seleceted_data['normalized_T'] = seleceted_data['normalized_T']
        m,n = seleceted_data.shape
        unique_values = np.unique(seleceted_data['num'])
        if len(unique_values) >= 3:
            model = Lmer(formula="lnR ~ normalized_T + (normalized_T||num)",
                             data=seleceted_data)
            model_result = model.fit()
            site.append(csv_file[0:6])
            igbp.append(site_IGBP[site_ID==csv_file[0:7]][0])
            T.append(Ti)
            Q10.append(np.exp(model_result['Estimate'][1]))
            SE.append(np.exp(model_result['Estimate'][1]+model_result['SE'][1])-np.exp(model_result['Estimate'][1]))
            p.append(model_result['P-val'][1])
            sig.append(model_result['Sig'][1])
            num.append(len(unique_values))
    result = np.array([site, igbp, T, Q10, SE, p, sig, num])
    result_pd = pd.DataFrame(result.T)
    save_path = './Q10_T_effect_nighttime.csv'
    result_pd.to_csv(save_path, index= 'a', header= False)
```
